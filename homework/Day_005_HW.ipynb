```python
def draw_histogram_by_column(data,column_name):
    """
    按照資料所指定的欄位名稱的對應資料，來畫直方圖
    Args
        - data: pandas的資料
        - column_name: 欄位名稱
    Return
        - None
    """
    samples=data[column_name].head(10)
    print(samples)
    plt.hist(samples)
    plt.show()
    
column_name="AMT_INCOME_TOTAL"
draw_histogram_by_column(app_train,column_name)
```
